```python
# ==============================================
# üìä GENERACI√ìN DE DATOS: CONSUMO ENERG√âTICO SEMANAL
# ==============================================
# Este script crea datos simulados para an√°lisis temporal.
# Se guardan autom√°ticamente en Google Drive:
# /Mi unidad/Ciencia-datos-2025-b-g1/FORO 1
# ==============================================

# ‚úÖ 1. Conectar con Google Drive
from google.colab import drive
drive.mount('/content/drive')

# ‚úÖ 2. Crear carpeta destino si no existe
import os
ruta_base = "/content/drive/MyDrive/Ciencia-datos-2025-b-g1/FORO 1"
os.makedirs(ruta_base, exist_ok=True)

# ‚úÖ 3. Importar librer√≠as
import pandas as pd
import numpy as np

# ‚úÖ 4. Generar datos simulados de consumo energ√©tico semanal
np.random.seed(42)
fechas = pd.date_range(start="2023-01-01", periods=104, freq="W")  # 2 a√±os de datos
# consumo base con patr√≥n estacional
consumo = 500 + 30 * np.sin(np.linspace(0, 3*np.pi, 104)) + np.random.normal(0, 10, 104)

# Crear DataFrame
df = pd.DataFrame({
    'fecha': fechas,
    'consumo_MWh': consumo
})

# ‚úÖ 5. Guardar los datos en CSV dentro de tu carpeta
ruta_datos = os.path.join(ruta_base, "consumo_energetico.csv")
df.to_csv(ruta_datos, index=False)

print(f"‚úÖ Archivo guardado correctamente en:\n{ruta_datos}")

# ‚úÖ 6. Vista previa de los datos
df.head()
```
